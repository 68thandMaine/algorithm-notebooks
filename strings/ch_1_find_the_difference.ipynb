# Find the Difference


[LeetCode](https://leetcode.com/problems/find-the-difference/)

### Problem Statement

Given two strings `s` and `t`, where `t` is generated by randomly shuffling the characters of `s` and adding one additional character, return the additional character added to `t`.

### Approach Overview

The solution leverages a character counting technique to identify the extra character:
1. Create a frequency map of characters in the first string
2. Iterate through the second string
3. Decrement character counts
4. Identify the character with a remaining count or the first unique character

### Key Insights
- Useful for understanding character frequency manipulation
- Demonstrates efficient string comparison techniques
- Multiple solution strategies exist (counting, sum difference, bitwise XOR)


```typescript
const count: Record<string, number> = {};
```

### Character Frequency Counting

Iterate through the first string to build a frequency map:
- Use an object to track character occurrences
- Increment the count for each character
- Handles repeated characters efficiently

```typescript
const s = 'aa';
for(const c of s) {
    count[c] = (count[c] || 0) + 1;
}

console.log(count);
```

### Identifying the Extra Character

Iterate through the second string to find the additional character:
- Check if the character exists in the frequency map
- Decrement the count for each character
- Identify the first character with:
  1. No existing count
  2. A count of zero
- This reveals the extra character added to the second string

#### Key Steps
- Traverse the second string systematically
- Use frequency map to track character occurrences
- Efficiently handle repeated characters
- Pinpoint the unique character

```typescript
const t = 'aaa'; // s randomized with "e" added

for(const c of t) {
    if(!count[c]) {
        console.log(c);
    }
      count[c]--;
}
```
